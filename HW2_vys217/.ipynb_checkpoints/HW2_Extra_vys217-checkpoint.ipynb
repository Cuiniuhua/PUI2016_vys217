```python
file_df = pd.read_csv("/gws/open/dataNYgov/OfficeAttorneyGeneral/ConsumerComplaint/rows.csv", 
                      parse_dates = ['Date received'], infer_datetime_format = True)
```

```python
 
file_df = file_df[['Date received','Complaint ID']]
file_df.columns = ['Date_received','Complaint_ID']
file_df.head(10)
```

```python
f = pd.value_counts(file_df.Date_received)
f= pd.DataFrame(f)
f['date'] = f.index
```

```python
f.reset_index(drop=True, inplace=True)
f.columns = ['No_Complaints','Date_received']
f.head(10)
```

```python
f_plot = f.plot(x='Date_received', y= 'No_Complaints',style='.', title='No. of Participants per Jurisdiction')
```
